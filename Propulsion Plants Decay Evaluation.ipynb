# Task 2 -Predicting GT Turbine decay state coefficient 

```python

cols_to_keep=['Lever position (lp) [ ]', 'Ship speed (v) [knots]',
       'Gas Turbine shaft torque (GTT) [kN m]',
       'Gas Turbine rate of revolutions (GTn) [rpm]',
       'Gas Generator rate of revolutions (GGn) [rpm]',
       'Starboard Propeller Torque (Ts) [kN]',
       'Port Propeller Torque (Tp) [kN]',
       'HP Turbine exit temperature (T48) [C]',
       'GT Compressor outlet air temperature (T2) [C]',
       'HP Turbine exit pressure (P48) [bar]',       
       'GT Compressor outlet air pressure (P2) [bar]',
       'Gas Turbine exhaust gas pressure (Pexh) [bar]',
       'Turbine Injecton Control (TIC) [%]', 'Fuel flow (mf) [kg/s]',
        'GT Compressor decay state coefficient.'
       ]
```

#Removing GT Compressor inlet air temperature (T1) [C],GT Compressor inlet air pressure(p2) due to very poor coorelation values

```python
X=df[cols_to_keep]
y=df['GT Turbine decay state coefficient.']
```

```python
x_train_turbine,x_test_turbine,y_train_turbine,y_test_turbine=train_test_split(X,y)
```

```python
linear_regressor_turbine=LinearRegression().fit(x_train_turbine,y_train_turbine)
```

```python
print('Training accuracy score: {}'.format(linear_regressor_turbine.score(x_train_turbine,y_train_turbine)))

print('Testing accuracy score: {}'.format(linear_regressor_turbine.score(x_test_turbine,y_test_turbine)))
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
linear_regressor_turbine_prediction=linear_regressor_turbine.predict(x_test_turbine)
rmse_linear_turbine=sqrt(mean_squared_error(y_test_turbine,linear_regressor_turbine_prediction))
print('Linear Regression rmse error:{}'.format(rmse_linear_turbine))
```

```python
random_forest_turbine=RandomForestRegressor(n_estimators=400,max_depth=40).fit(x_train_turbine,y_train_turbine)
```

```python
print('Training accuracy score: {}'.format(random_forest_turbine.score(x_train_turbine,y_train_turbine)))
print('Testing accuracy score: {}'.format(random_forest_turbine.score(x_test_turbine,y_test_turbine)))
```

#Further Minimizing rmse loss using Random Forests

```python

random_forest_turbine_prediction=random_forest_turbine.predict(x_test_turbine)
rmse_rf_turbine=sqrt(mean_squared_error(y_test_turbine,random_forest_turbine_prediction))
print('Random Forest  rmse error:{}'.format(rmse_rf_turbine))
```

```python
import matplotlib.pyplot as plt
feature_importance = random_forest_turbine.feature_importances_
sorted_idx = np.argsort(feature_importance)
pos = np.arange(sorted_idx.shape[0]) + .5
print(pos)
plt.figure()
plt.barh(pos, feature_importance[sorted_idx], align='center')
plt.yticks(pos, np.array(x_train_turbine.columns)[sorted_idx])
plt.title('Feature Importance(Random Forest Regressor)')
```

#Trying to further minimize rmse with a deep learning model

```python

import tensorflow as tf
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model_turbine=Sequential([Dense(128,activation='relu',input_shape=(x_train_turbine.shape[1],)),
                      Dense(128,activation='relu'),
                     Dense(128,activation='relu'),
                     Dense(128,activation='relu'),
                     Dense(64,activation='relu'),
                     Dense(64,activation='relu'),
                     Dense(1)])
```

```python
model_turbine.summary()
```

```python
model_turbine.compile(optimizer='adam',loss='mse',metrics=['mse'])
```

```python
history=model_turbine.fit(x_train_turbine,y_train_turbine,validation_split=0.15,epochs=30,batch_size=64)
```

```python
evaluate=model_turbine.evaluate(x_test_turbine,y_test_turbine,verbose=2)
```

```python
rmse_dl=sqrt(evaluate[1])
print(rmse_dl)
```

#Hence the random forest model was the most effective in reducing rmse
